```python
def try_summarize(x, ratio):
    try:
        return(summarize(x,ratio=ratio))
    except:
        return('')
sub_complaints_data_['TextRank Summary'] = sub_complaints_data_['Consumer complaint narrative'].\
apply(lambda x: try_summarize(x,ratio=0.20))
```
